```python
import pandas as pd
```

```python
file_name = 'stm_date만_topic.csv'
a = file_name.split('_')[-1].split('.')[0]
file_name = '보건복지_label.csv'
a = file_name.split('_')[-1].split('.')[0]
```

```python
csv = pd.read_csv(file_name, encoding='cp949')
csv.head(2)
```

```python
monthpertopic = {}
for i in range(0, len(csv)):
    date = csv.loc[i]['covariate.value']
    topic = csv.loc[i]['topic']
    estimate = csv.loc[i]['estimate']
    if date in monthpertopic:
        if topic in monthpertopic[date]:
            a = 1
        else:
            monthpertopic[date][topic] = estimate
    else:
        tmp = {}
        tmp[topic] = estimate
        monthpertopic[date] = tmp
```

```python
monthpertopic
```

```python
a = []
for key, val in monthpertopic.items():
    for key2, val2 in val.items():
        if key2 == 9 or key2 == 5 or key2 == 11:
            tmp = [key, key2, val2]
            
            # v = (int(key[:4]) - 2020) * 12 + (int(key[4:6]))
            v = str(2020 + ((key-1) // 12)) + '-' + '%02d' %((key-1)%12 + 1) + '-' + '01'
            tmp.append(v)
            # dd = key[:4]+ '-' + key[4:6] + '-' + '01'
            # tmp.append(dd)
            a.append(tmp)
```

```python
a
```

```python
c = ['month', 'topicnum', 'est', 'date']
df_list = pd.DataFrame(a, columns=c)
```

```python
df_list
```

```python

df_list = df_list.sort_values(by=['date'], axis=0)
```

```python
import numpy as np
df_list['est'] = np.exp(np.log(df_list['est']) + 10)
```

```python
import plotly.express as px

fig = px.line(df_list, x="date", y="est", color='topicnum', range_x=['2019-12-01','2022-01-01'])
fig.update_layout(title='5,9,11 보건복지',paper_bgcolor='#f8f8f8')
    
fig.show()
fig.write_html("./5911file.html")
```

```python
import plotly.express as px

fig = px.scatter(df_list, x="date", y="topicnum",
	         size="est", color="topicnum",
                 hover_name="topicnum", log_x=False, size_max=13
                )
fig.update_layout(
    title='전체 토픽별 빈도')
fig.show()
fig.write_html("./file.html")
```

```python
import plotly.express as px
for i in range(1, 26):
    a = []
    for key, val in monthpertopic.items():
        for key2, val2 in val.items():
            if key2 == i:
                tmp = [key, key2, val2]
                key = str(key)
                # v = (int(key[:4]) - 2020) * 12 + (int(key[4:6]))
                v = key[:4]+ '-' + key[4:6] + '-' + '01'
                tmp.append(v)
                a.append(tmp)
    c = ['month', 'topicnum', 'est', 'index']
    df_list = pd.DataFrame(a, columns=c)
    df_list['est'] = np.exp(np.log(df_list['est']) + 10)
    df_list = df_list.sort_values(by=['index'], axis=0)
    fig = px.line(df_list, x="index", y="est", title='남궁다', range_x=['2019-12-01','2022-01-01'])
    fig.update_layout(
        title=' ')
    # fig.show()
    fig.update_xaxes(showgrid=False)
    fig.update_yaxes(showgrid=False)
    fig.write_html("./a" + str(i) +"file.html")
```
